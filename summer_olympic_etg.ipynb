```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Reading the Data**


```python
df = pd.read_csv("summer.csv")
df
```

```python
print("Number of cities in which Summer Olympics were held  so far are:" ,len(df['City'].unique()))
```

**2. Which sport is having most number of Gold Medals so far?**

```python
data = df.values
lst = []
for Sport in df['Sport'].unique():
  var = 0
  for i in range(len(data)):
      if (data[i][2] == Sport and data[i][8] == 'Gold'):
        var += 1
  lst.append([Sport,var])
lst
```

```python
no = []
for i in range(0,len(lst)):
  no.append(lst[i][1])
for i in range(0,len(lst)):
  if (max(no)==lst[i][1]):
    spt = lst[i][0]
    break
  break
print("Sport which is having most number of Gold Medals so far:",spt)
print("Number of Gold Medals",spt,"sport is having are:",max(no))
```

**Graphical Analysis of Sports having Gold Medal**

```python
pd.DataFrame(lst,columns = ["Sport","Number of Gold Medals"]).sort_values(by='Sport', ascending= True).plot(x = 'Sport', y = 'Number of Gold Medals', kind = 'bar', figsize = (20,5))
```

**Graphical Analysis of Top 5 Sports having Gold Medal**

```python
gd = df[df['Medal'] == 'Gold']
gd = gd.groupby('Sport').count()['Medal'].sort_values(ascending = False).head()
print(gd)
gd.plot.bar(figsize =(10,5))
```

**3. Which sport is having most number of medals so far?**

```python
data = df.values
md = []
for Sport in df['Sport'].unique():
  ct = 0
  for i in range(len(data)):
      if (data[i][2] == Sport and data[i][8] != None):
        ct += 1
  md.append([Sport,ct])
md
```

```python
nu = []
for i in range(0,len(md)):
  nu.append(md[i][1])
for i in range(0,len(md)):
  if (max(nu)==md[i][1]):
    spot = md[i][0]
    break
  break
print("Sport which is having most number of Medals so far:",spot)
print("Number of Medals",spot,"sport is having are:",max(nu))
```

**Graphical Analysis of Sports having most number of Medals**

```python
pd.DataFrame(md,columns = ["Sport","Number of Medals"]).sort_values(by='Sport', ascending= True).plot(x = 'Sport', y = 'Number of Medals', kind = 'bar', figsize = (20,5))
```

**4. Which player has won most number of medals?**

```python
data = df.values
pl = []
for Athlete in df['Athlete'].unique():
  cn = 0
  for i in range(len(data)):
      if (data[i][4] == Athlete and data[i][8] != None):
        cn += 1
  pl.append([Athlete,cn])
pl
```

```python
nm = []
for i in range(0,len(pl)):
  nm.append(pl[i][1])
```

```python
for i in range(len(pl)):
  if (max(nm) == pl[i][1]):
    print("Player who has won most number of medals:",pl[i][0])
    print("Number of Medals",pl[i][0],"has won:",max(nm))
```

**Graphical Analysis of Top 5 Players and number of Medals they won**

```python
pler = df.groupby('Athlete').count()['Medal'].sort_values(ascending=False).head()
print(pler)
pler.plot.bar(figsize= (10,5))
```

**6. In which year India won first Gold Medal in Summer Olympics?**

```python
player=df[df['Medal']=='Gold']
yr=player.sort_values(['Year'],ascending=True)

gld=yr[yr['Country']=='IND'].iloc[0]

print("The year in which India won it's First Gold Medal in Summer Olympics is ",gld['Year'] ,"\n")
print(gld)
```
